## 4.6 Combining & Exporting Data

### This script contains the following points:
#### 1. Import data
#### 2. Merge Datasets
#### 3. Drop columns
#### 4. Export data in pickle format

```python
# Importing libraries

import pandas as pd
import numpy as np
import os
```

#### 1. Import data

```python
# Define file path
path = '/Users/mariasampogna/Documents/Career Foundry/Data Analytics Immersion/Achievement 4/Instacart Basket Analysis'
```

```python
# Import ords_prioir dataset

df_ords_prior = pd.read_csv(os.path.join(path, '02 Data', 'Original Data', 'order_products__prior.csv'),index_col = False)
```

```python
# Import ords dataset

df_ords = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'orders_checked.csv'), index_col = False)
```

```python
# Check the output

df_ords_prior.head()
```

```python
df_ords.head()
```

#### 2. Merge datasets

```python
# Execute merge code

df_merged_large = df_ords.merge(df_ords_prior, on = 'order_id', indicator = True)
```

```python
# Check the output

df_merged_large.head()
```

```python
df_merged_large['_merge'].value_counts()
```

```python
# Check merge rate with outer merge

df_merged_large_outer = df_ords.merge(df_ords_prior, on = 'order_id', how = 'outer', indicator = True)
```

```python
df_merged_large_outer['_merge'].value_counts()
```

#### 3. Drop columns

```python
# Deop unnamed column from merged dataset

df_merged_large.drop(columns=['Unnamed: 0'])
```

#### 4. Export data in pickle format

```python
# export merged dataset to pickle file

df_merged_large.to_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_combines.pkl'))
```
