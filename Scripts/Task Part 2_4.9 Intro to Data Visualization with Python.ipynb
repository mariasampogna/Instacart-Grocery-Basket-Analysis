#### 02. Create a Bar Chart for order_day_of_week

```python
# Create a clean bar chart without colors

bar = ords_prods_custs['order_day_of_week'].value_counts().plot.bar()
```

```python
# Export clean bar chart

bar.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'bar_orders_dow_clean.png'))
```

#### 03. Create a Histogram

```python
# Create a histogram of the “order_hour_of_day” column

hist = ords_prods_custs['order_hour_of_day'].plot.hist(bins = 24)
```

This histogram is showing how often each hour of the day is associated with an order in the datframe. This shows us that most orders are happening in the middle of the day, between about 9 AM and 4 PM.

#### 04. Create a Bar Chart for loyalty_flag

```python
# Create a bar char for loyalty_flag

bar_loyalty = ords_prods_custs['loyalty_flag'].value_counts().plot.bar()
```

#### 05. Create a Data Sample

```python
# Create a list holding True/False value to the test np.random.rand() <=0.7

np.random.seed(4)
dev = np.random.rand(len(ords_prods_custs)) <= 0.7
```

```python
# Store 70% of the sample in the dataframe big

big = ords_prods_custs[dev]
```

```python
# Store 30% of the sample in the dataframe small

small = ords_prods_custs[~dev]
```

```python
# check the lengths of the outputs

len(ords_prods_custs)
```

```python
len(big) + len(small)
```

#### 06. Create a Line Chart of prices and order_hour_of_day

```python
# Create dataframe of columns necessary for line chart

df_2 = small[['order_hour_of_day','prices']]
```

```python
# Check the output

df_2.head()
```

```python
# create a line chart of prices by hour of day

line = sns.lineplot(data = df_2, x = 'order_hour_of_day',y = 'prices')
```

#### 07. Create a Line Chart of age and n_dependants

```python
# Create dataframe of columns necessary for line chart

df_3 = small[['age','n_dependants']]
```

```python
# Check the output

df_3.head()
```

```python
# create a line chart of number of dependants by age

line_2 = sns.lineplot(data = df_3, x = 'age',y = 'n_dependants')
```

This line chart shows that there does not appear to be a correlation between age and number of dependents. There is no trend showing that as age increase or decreases, number of dendents increase or decreases.

#### 08. Create a Scatterplot

```python
# Create a scatterplot of age and income

scatter = sns.scatterplot(data = ords_prods_custs, x = 'age',y = 'income')
```

This scatterplot shows that income tends to be distributed in the lower income ranges across all age groups. We can also see that above age 40, income is higher than in younger ages. 

Up to age 40, income is mainly distributed beween 0 and 200,000, with more dispersion in higher ranges up to about $400,000.

Over age 40, income is mainly distributed between 0 and 300,000, with more dispersion in higher ranges up to about $600,000
