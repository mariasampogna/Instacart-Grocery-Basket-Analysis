#### 03. Data Wrangling

```python
# Rename columns for consistency

cust.rename(columns = {'First Name' : 'first_name', 'Surnam' : 'last_name', 'Gender' : 'gender', 'STATE' : 'state', 'Age' : 'age', 'fam_status' : 'marital_status'}, inplace = True)
```

```python
# Check the output

cust.head()
```

#### 04. Data Consistency Checks

```python
# Check for mixed type columns

for col in cust.columns.tolist():
  weird = (cust[[col]].applymap(type) != cust[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (cust[weird]) > 0:
    print (col)
```

```python
# Change data type for first_name to string

cust['first_name'] = cust['first_name'].astype('str')
```

```python
# Check for missing values

cust.isnull().sum()
```

```python
# Check for full duplicates

df_dups = cust[cust.duplicated()]
```
